```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)

#df = pd.read_csv("data/^990100-USD-STRD (1).csv")
df = pd.read_csv("data/^SPX (1).csv")

df = df.dropna(axis=0)

df["Date"] = pd.to_datetime(df["Date"])
df = df[df["Date"] >= "1994-03-08"]
#df.to_csv("data/SP500_Tobi_30j.csv")

df
```

```python
def calc_returns(prices):
    returns = [0]
    counter = len(prices)
    for i in range(1, counter):
        ret = prices[i] / prices[i-1] - 1
        returns.append(ret)
    return returns

df["daily_movement"] = calc_returns(df["Adj Close"].to_list())
df["daily_movement_2x"] = df["daily_movement"] * 2
df["daily_movement_3x"] = df["daily_movement"] * 3
df
```

```python
# cum returns
def cum_returns(returns):
    cum_returns = [1]
    counter = len(returns)
    for i in range(1, counter):
        ret = (returns[i] + 1) * cum_returns[i-1]
        cum_returns.append(ret)
    return cum_returns

df["cum_return"] = cum_returns(df["daily_movement"].to_list())
df["cum_return_2x"] = cum_returns(df["daily_movement_2x"].to_list())
df["cum_return_3x"] = cum_returns(df["daily_movement_3x"].to_list())
df
```

```python
# adj close
df["adj_close_2x"] = df["cum_return_2x"] * df["Adj Close"].to_list()[0]
df["adj_close_3x"] = df["cum_return_3x"] * df["Adj Close"].to_list()[0]
df["adj_close"] = df["cum_return"] * df["Adj Close"].to_list()[0]
df
```

```python
# moving average
from statistics import mean

def moving_average(data, days):
    avg = [np.NaN] * days
    data = data.to_list()
    
    for i in range(days, len(data)):
        subset = data[i-days:i]
        avg.append(mean(subset))

    return avg

df["mavg_300_adj_close"] = moving_average(df.adj_close, 300)
df["mavg_300_adj_close_2x"] = moving_average(df.adj_close_2x, 300)
df["mavg_300_adj_close_3x"] = moving_average(df.adj_close_3x, 300)
df["mavg_200_adj_close"] = moving_average(df.adj_close, 200)
df["mavg_200_adj_close_2x"] = moving_average(df.adj_close_2x, 200)
df["mavg_200_adj_close_3x"] = moving_average(df.adj_close_3x, 200)
df["mavg_100_adj_close"] = moving_average(df.adj_close, 100)
df["mavg_100_adj_close_2x"] = moving_average(df.adj_close_2x, 100)
df["mavg_100_adj_close_3x"] = moving_average(df.adj_close_3x, 100)
df.dropna(axis=0)
```

```python
"""
Investment Strategy: Sell when adj_close < moving average, buy when adj_close > moving average
"""

df = df.dropna(axis=0)

def apply_strategy(start_amount, adj_close, mavg, returns, ter, tax=0.25*0.7, transaction_cost=1, spread=0.02):
    operations = ["buy"]
    operations_balance = [start_amount]
    balance = []
    adj_close = adj_close.to_list()
    mavg = mavg.to_list()
    returns = returns.to_list()
    
    for i in range(len(adj_close)):
        # claculate return for the day if in the market
        if operations[-1] == "buy":
            start_amount = start_amount * (1 + returns[i])

        balance.append(start_amount)

        print("Amount:", start_amount, "Operation:", operations[-1], "Price:", adj_close[i], "AVG:", mavg[i])

        # buy, if not in the market and adj close price rises above avg
        if adj_close[i] > mavg[i] and operations[-1] != "buy":
            operations.append("buy")
            operations_balance.append(start_amount)
            # transaction_cost
            start_amount = start_amount - transaction_cost
        # sell, if in the maket and dj close price falls below avg
        elif adj_close[i] < mavg[i] * 1 and operations[-1] != "sell":
            operations.append("sell")
            # tax
            if start_amount > operations_balance[-1]:
                start_amount = start_amount - (start_amount - operations_balance[-1]) * tax
            # spread
            start_amount = start_amount * (1 - spread)
            # transaction_cost
            start_amount = start_amount - transaction_cost

        # TER
        if i % 200 == 0:
            start_amount = start_amount * (1 - ter)

    print(len(operations))
    print(start_amount)
    return balance
        
amount = df["adj_close_3x"].to_list()[0]
balance_100 = apply_strategy(amount, df.adj_close_3x, df.mavg_100_adj_close_3x, df.daily_movement_3x, 0.0075)
balance_200 = apply_strategy(amount, df.adj_close_3x, df.mavg_200_adj_close_3x, df.daily_movement_3x, 0.0075)
balance_300 = apply_strategy(amount, df.adj_close_3x, df.mavg_300_adj_close_3x, df.daily_movement_3x, 0.0075)
base = apply_strategy(amount, df.adj_close, df.adj_close, df.daily_movement, 0.001)
comp = pd.DataFrame({"avg 100": balance_100, "avg 200": balance_200, "avg 300": balance_300, "base (wo trading and leverage)": base})
comp = comp.reset_index()
comp["date"] = df["Date"].to_list()
comp["mavg_200_adj_close_3x"] = df["mavg_200_adj_close_3x"].to_list() #/ df["adj_close_3x"].to_list()[0] * amount
comp["mavg_200_adj_close_3x"] = comp["mavg_200_adj_close_3x"] / df["adj_close_3x"].to_list()[0] * amount
comp["adj_close_3x"] = df["adj_close_3x"].to_list()

fig = px.line(comp, x="date", y=["avg 100", "avg 200", "avg 300", "base (wo trading and leverage)", "mavg_200_adj_close_3x", "adj_close_3x"])
fig.write_html('figures/first_figure4.html', auto_open=True)
comp.to_csv("test.csv")
```
