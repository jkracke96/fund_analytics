```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)

#df = pd.read_csv("data/^990100-USD-STRD (1).csv")
df = pd.read_csv("data/^SPX (1).csv")

df = df.dropna(axis=0)

df["Date"] = pd.to_datetime(df["Date"])
df = df[df["Date"] >= "1994-03-08"]
#df.to_csv("data/SP500_Tobi_30j.csv")

df
```

```python
def calc_returns(prices):
    returns = [0]
    counter = len(prices)
    for i in range(1, counter):
        ret = prices[i] / prices[i-1] - 1
        returns.append(ret)
    return returns

df["daily_movement"] = calc_returns(df["Adj Close"].to_list())
df["daily_movement_2x"] = df["daily_movement"] * 2
df["daily_movement_3x"] = df["daily_movement"] * 3
df
```

```python
# cum returns
def cum_returns(returns):
    cum_returns = [1]
    counter = len(returns)
    for i in range(1, counter):
        ret = (returns[i] + 1) * cum_returns[i-1]
        cum_returns.append(ret)
    return cum_returns

df["cum_return"] = cum_returns(df["daily_movement"].to_list())
df["cum_return_2x"] = cum_returns(df["daily_movement_2x"].to_list())
df["cum_return_3x"] = cum_returns(df["daily_movement_3x"].to_list())
df
```

```python
# moving average
from statistics import mean

def moving_average(data, days):
    avg = [np.NaN] * days
    data = data.to_list()
    
    for i in range(days, len(data)):
        subset = data[i-days:i]
        avg.append(mean(subset))

    return avg

#df["mavg_300_adj_close"] = moving_average(df.adj_close, 300)
#df["mavg_300_adj_close_2x"] = moving_average(df.adj_close_2x, 300)
#df["mavg_300_adj_close_3x"] = moving_average(df.adj_close_3x, 300)
df["mavg_200_adj_close"] = moving_average(df.adj_close, 200)
df["mavg_200_adj_close_2x"] = moving_average(df.adj_close_2x, 200)
df["mavg_200_adj_close_3x"] = moving_average(df.adj_close_3x, 200)
df["mavg_100_adj_close"] = moving_average(df.adj_close, 100)
df["mavg_100_adj_close_2x"] = moving_average(df.adj_close_2x, 100)
df["mavg_100_adj_close_3x"] = moving_average(df.adj_close_3x, 100)
df.dropna(axis=0)
```
